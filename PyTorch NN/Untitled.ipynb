```python
import torch
from torch.autograd import Variable
import matplotlib.pyplot as plt
import torch.optim as optim
import numpy as np
```

```python
x_variables =  np.array([ [1.,1.,1.],[1.,0.,1.],[0.,0.,1.],[0.,1.,1.] ])
x = Variable(torch.FloatTensor(x_variables))
print(x)
y_variables = np.array([1,0,0,1])
y = Variable(torch.FloatTensor(y_variables))
print(y)
```

```python
N, D_in, H, D_out = len(x_variables), len(x_variables[0]), 3, 1
# N = number of things to classify
# D_in = flattened image or long feautre array
# H = number of nodes on hidden layer
# D_out = number of output nodes (classification nodes)
```

```python
x = Variable(torch.randn(N, D_in))
print(x)
```

```python
y = Variable(torch.randn(N, D_out), requires_grad=False)
print(y)

learning_rate = 1e-3
```

```python
model = torch.nn.Sequential(
torch.nn.Linear(D_in,H), # D_in goes into H
torch.nn.ReLU(), #Pass through non linear func
torch.nn.Linear(H,D_out), # H goes into D_out
)
```

```python
optimizer = optim.Adam(model.parameters(),lr=learning_rate)
```

```python
loss_fn = torch.nn.MSELoss(size_average=False)
```

```python
learning_rate = 1e-3
```

```python
plt.figure()
for t in range(1000):
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(t, loss.data[0])
    plt.plot(t,loss.data[0],'k.')
    optimizer.zero_grad() #zero gradients before backwards prop
    loss.backward() #Back prop
    for param in model.parameters():
        param.data -= learning_rate * param.grad.data
    
```

```python
plt.show()
```

```python
print(model.parameters)
```

```python
class TwoLayerNet(torch.nn.Module):
    def __init__(self,D_in,H,D_out):
        ### Constructor ###
        super(TwoLayerNet,self).__init__()
        self.linear1 = torch.nn.Linear(D_in,H)
        self.linear2 = torch.nn.Linear(H,H)
        self.linear3 = torch.nn.Linear(H,D_out)
        ###    ###   ###  ###
        
    def forward(self,x):
        h_relu = self.linear1(x).clamp(min=0) #stops anything being below 0 (definition of a relu func)
        out_1 = self.linear2(h_relu).clamp(min=0)
        y_pred = self.linear3(out_1)
        return y_pred
```

```python
model = TwoLayerNet(D_in, H, D_out)
```

```python
x_variables =  np.array([ [1.,1.,1.],[1.,0.,1.],[0.,0.,1.],[0.,1.,1.] ])
x = Variable(torch.FloatTensor(x_variables))
print(x)
y_variables = np.array([1,0,0,1])
y = Variable(torch.FloatTensor(y_variables))
print(y)
```
